# Selenium

```python
import requests
from selenium import webdriver
from time import sleep
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#Creating list of urls
url_list = []
for year in range(2014,2022):
    for month in range (1,13):
        url_list.append(f"https://www.wunderground.com/history/monthly/EHAM/date/{year}-{month}")
```

```python
url_list = url_list[:5]
```

```python
#Opciones de Selenium para evitar que sepan que eres un robot
opciones= Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
#para ocultarme como robot, pero me sigue diciendo que hay un software automatico usando chrome
opciones.add_experimental_option('useAutomationExtension', False)
opciones.add_argument('--start-maximized') #empezar maximizado
opciones.add_argument('user.data-dir=selenium') #guarda las cookies
opciones.add_argument('drive/adblock.crx') #adblocker
opciones.add_argument('--incognito')#incognito window
```

```python
#connect with chrome
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
results_list = []    
for i in url_list:   
    #Go to URL
    driver.get(i)
    # After coping the full X-path in the webpage
    sleep(5)
    my_info = driver.find_element_by_xpath("/html/body/app-root/app-history/one-column-layout/wu-header/sidenav/mat-sidenav-container/mat-sidenav-content/div/section/div[2]/div[1]/div[5]/div[1]/div/lib-city-history-observation/div/div[2]")
    print("Done")
    print(my_info)
    results_list.append(my_info.text) #we have to append the result.text because if we try to extract the text afterwars it doesn't work
driver.quit()
```

## Limpieza y creación de dataframe completo

### Create list of month,year

```python
month_year = []
for year in range(2014,2022):
    for month in range (1,13):
        month_year.append([month,year])
        
```

```python
month_year[:3]
```

Create new dataframe from list of results 

```python
clima_df = pd.DataFrame(columns = columnas)
for i in range (len(df)):
    x = list(df.iloc[i][0].split("\n"))
    mi_ind = x[1:].index('Max Avg Min')
    single_df = pd.DataFrame(np.array(x[1:]).reshape(7,mi_ind)).T
    single_df.columns = ["Time", "Temperature (° F)", "Dew Point (° F)", "Humidity (%)", "Wind Speed (mph)", "Pressure (Hg)", "Precipitation (in)"]
    single_df['month'] = month_year[i][0]
    single_df['year'] = month_year[i][1]
    clima_df = pd.concat([clima_df, single_df], axis=0)
```

### Columns expand

Como nuestras columnas tienen 3 columnas dentro realmente hacemos un expand

```python
clima_df[['Tmax','Tavg','Tmin']] = clima_df["Temperature (° F)"].str.split(" ",expand=True)
```

```python
clima_df[['DPmax','DPavg','DPmin']] = clima_df["Dew Point (° F)"].str.split(" ",expand=True)
```

```python
clima_df[['Humidity_max','Humidity_avg','Humidity_min']] = clima_df["Humidity (%)"].str.split(" ",expand=True)
```

```python
clima_df[['wind_max','wind_avg','wind_min']] = clima_df["Wind Speed (mph)"].str.split(" ",expand=True)
```

```python
clima_df[['pressure_max','pressure_avg','pressure_min']] = clima_df["Pressure (Hg)"].str.split(" ",expand=True)
```

```python
clima_df = clima_df.rename(columns={'Time': 'day'})
```

```python
clima_df.head()
```

Como aún no hemos hecho `reste index`todas las cabeceras tienen index cero 

```python
clima_df = clima_df.drop([0])#quitamos todas la cabeceras porque tienen el index 0 
```

```python
clima_df.reset_index(drop=True)
```

Por último quitamos las columnas originales

```python
clima_df = clima_df.drop(["Temperature (° F)", "Dew Point (° F)", "Humidity (%)", "Wind Speed (mph)", "Pressure (Hg)"], axis = 1)
```
